```python
import numpy as np
```

Step 1-Create Sigmoid Function and apply array to it.

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

z=np.array([-2,-1,0,1,2])
print(sigmoid(z))
```

Step 2-Visualize Sigmoid using array of range [-10,10]

```python
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1/(1+np.exp(-x))

range_a=np.arange(-10,10)

plt.title('Sigmoid of range [-10,10]')
plt.xlabel('range')
plt.ylabel('sigmoid(range)')
plt.xticks([-10,0,10],['-10','0','10'])
x=np.linspace(range_a,10)
plt.plot(x,sigmoid(x),'+')
plt.show()
```
